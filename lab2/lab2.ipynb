```julia
using CSV
input="data1.csv"
mydata1=CSV.read(input, delim=";")
```

```julia
using CSV
input="data2.csv"
mydata2=CSV.read(input, delim=";")
```

```julia
 dataset1 = by(mydata1, :size) do df
          DataFrame(m = mean(df.time), v = var(df.time))
       end
```

```julia
dataset2 = by(mydata2, :size) do df
          DataFrame(m = mean(df.time), v = var(df.time))
       end
```

```julia
using Plots
x = dataset1[:size]
m1 = dataset1[:m]
m2 = dataset2[:m]
v1 = dataset1[:v]
v2 = dataset2[:v]
scatter(x, [m1, m2], colour = [:red :green], label=["Mean 1" "Mean 2"], yerr= [v1 v2], layout =2)
```
