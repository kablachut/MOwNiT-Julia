```julia
using Roots  
using Plots
using ForwardDiff
using DataFrames 
```

```julia
f(x) = sin(x) - x / 2
D(x) = x->ForwardDiff.derivative(f, float(x))

println("Function:  sin(x) - x / 2 ")
x = find_zero(f, (-2, 2), FalsePosition(), verbose=true)
println(x)
println(iszero(x))
y = find_zero((f, D(f)), 1, Roots.Newton(),verbose=true)
println(y)
println(iszero(y))
z = find_zero(f, 1, Order2(), verbose=true)
println(z)
println(iszero(z))
```
